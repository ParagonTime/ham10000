```python
import torch
import random
import numpy as np
import pandas as pd
import torchvision.datasets
```

```python
y_test = torch.load('gdrive/MyDrive/datasets/ham_100_y_test.pt')
```

```python
ResN = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ansambl_ham10000_resnet.csv')
ResN = ResN.drop(['Unnamed: 0'], axis=1)
```

```python
DensN = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ansambl_ham10000_densenet.csv')
DensN = DensN.drop(['Unnamed: 0'], axis=1)
```

```python
ResNX = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/ansambl_ham10000_resnext.csv')
ResNX = ResNX.drop(['Unnamed: 0'], axis=1)
```

```python
ADAM = pd.DataFrame()

ADAM['ResNet'] = ResN['Adam']
ADAM['ResNeXt'] = ResNX['Adam']
ADAM['DensN'] = DensN['Adam']

data = np.zeros((2003,7), dtype=int)
ARes = pd.DataFrame(data)

for i in range(2003):
    for j in range(3):
        if j == 0:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.7693
        if j == 1:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.7913
        if j == 2:
            ARes.iloc[i, ADAM.iloc[i, j]] += 0.8247


anslist = ARes.idxmax(axis=1)
accuracy = (torch.tensor(anslist) == y_test).float().mean().item()
print('Adam:', accuracy)
```

```python
PNMBelief = pd.DataFrame()

PNMBelief['ResNet'] = ResN['PNMBelief']
PNMBelief['ResNeXt'] = ResNX['PNMBelief']
PNMBelief['DensN'] = DensN['PNMBelief']

data = np.zeros((2003,7), dtype=int)
ARes = pd.DataFrame(data)

for i in range(2003):
    for j in range(3):
        if j == 0:
            ARes.iloc[i, PNMBelief.iloc[i, j]] += 0.82126
        if j == 1:
            ARes.iloc[i, PNMBelief.iloc[i, j]] += 0.81377
        if j == 2:
            ARes.iloc[i, PNMBelief.iloc[i, j]] += 0.8522


anslist = ARes.idxmax(axis=1)
accuracy = (torch.tensor(anslist) == y_test).float().mean().item()
print('PNMBelief:', accuracy)
```
